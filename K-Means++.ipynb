### K-Means++

- K-Means is quite sensitive to initalization, if the init is not good, our algorithm is not able to make desired number
of clusters.
- To overcome this problem, we use technique called K-Means++ (described in paper [Robust Seed Selection for K-Means type of Algorithms](https://arxiv.org/abs/1202.1585))
    which chooses initial centers so that they are statiscallly close to final ones.

- The method is default choice for sklearn, so we will use it directly without going into much maths!

## Using Sci-kit Learn

```python
# sklarn.clusters.KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X,y) # computes the cluster centers and the training part is done
# it by default uses KMeans++ initialization
```

```python
centers = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(centers[:,0],centers[:,1],marker='*',color='red')
plt.show()
```

```python
# lets compute prediction
pred = kmeans.labels_
```

```python
print(pred)
# so we can see that it randomly assigns each cluster a label and then assigns 

# each point to a cluster ie predicts for every point the label of cluster to which the point belongs
```

```python
#  Giving coloring according to the labels
plt.scatter(X[:,0],X[:,1],c=y) # instead of c=y we could also write c=pred in this case as it would have assigned different color for different labels 
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black')
plt.title("Prediction")
plt.show()
```

```python
# Lets see the actual graph
plt.scatter(X[:,0],X[:,1],c=y) # instead of c=y we could also write c=pred in this case as it would have assigned different color for different labels 
plt.title("Actual")
plt.show()
```
