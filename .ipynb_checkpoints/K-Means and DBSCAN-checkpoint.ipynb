## Problem with K-Means
- can fail sometimes on complex datasets

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles,make_moons # they are non-linearly seperable datasets
```

```python
X,Y = make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
# so we can see that there are two clusters in our data which we want our algorithm to predict
# we want our each class to be predicted seprately by algorithm, and clearly its not a linearly seperable and neither it is a convex data unlike other examples we did earlier
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
km.fit(X)
```

```python
centers = km.cluster_centers_
label = km.labels_

plt.scatter(X[:,0],X[:,1],c=label) # c=label this means that our color is predicted by the labels 
plt.scatter(centers[:,0],centers[:,1],color='red')
plt.show()
```

So we can see that KMeans was not able to correctly predict the classes as it is a non-linear dataset,a dn such problems our quite difficult for the KMeans to solve.So Kmeans does fail on datasets which are quite complex and are not linear. SO to solve these type of algorithms we will learn one more algorithm called as 'DBSCAN'.

## DBSCAN

- <b>Density Based Spatial Clustering of Applications with Noise</b>
- Based upon the idea , that a cluster is a high density area surrounded by a low density region
- Starts by exploring a small area,if density is "good enough" it is considered as a part of cluster , and explores neighbours and so on,so it groups regions based upon density!

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN?
```

```python
dbs = DBSCAN(eps=0.2,min_samples= 5) # it does not have any parameter that can take total number of classes
```

```python
dbs.fit(X)
```

```python
Ypred = dbs.fit_predict(X)  # predicting the cluster labels of the clusters for each point in X 
```

```python
plt.scatter(X[:,0],X[:,1],c=Ypred)
```

SO now we can see that it performed quite weel in comparison to the KMeans clustering algorithm for the same non-linear ,complex dataset.In this algorithm we just need to play with the params eps and min_samples of DBSCAN object to get correct clusters for such a complex data
